```python
filename="b9_tv3_corpus.conllu"
file=file_to_conllu(filename)

def repeated_sentences_finder(filename):
    file= file_to_conllu(filename)

    l= [(sentence.metadata["custom_sent_id"], sentence.metadata["text"]) for sentence in file]
    l = sorted(l, key=lambda x: x[1])
    #sents= sorted({k: v for k, v in sorted(dic.items(), key=lambda item: item[1])})
    l_repeated = []
    c=0
    repetitions= {}
    for i, item in enumerate(l):
        #print(item)
        if l[i][1] == l[i-1][1]:
            #print(l[i], "\n", l[i-1])
            l_repeated.append(l[i])
            l_repeated.append(l[i-1])
            c = c+1
            pair = (l[i][0][:-4], l[i-1][0][:-4])
            if pair not in repetitions.keys():
                repetitions[pair] = 1
            else:
                repetitions[pair] = repetitions[pair] +1

    print(f"Total sents{len(l)}, repeated {c}, {100*c/len(l)} %")
    return l_repeated







l_repeated= repeated_sentences_finder(filename)
```
